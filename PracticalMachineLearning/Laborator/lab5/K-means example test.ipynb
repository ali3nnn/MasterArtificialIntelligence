```python
X_test = np.array([[1, 2], [1, 4], [1, 0],[4, 2], [4, 4], [4, 0]])
# plt.scatter(X_test[:,0], X_test[:,1])
X_test.shape
```

```python
wcss = []
for i in range(1, (len(X_test)+1)):
    print(i)
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=20, random_state=0)
    kmeans.fit(X_test)
    print(kmeans.inertia_)
    wcss.append(kmeans.inertia_)
    
```

```python
# print(wcss)
```

```python
plt.plot(range(1, (len(X_test)+1)), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=3, n_init=1, random_state=0)
pred_y = kmeans.fit_predict(X_test)
```

```python
for i in range(len(X_test)):
    print(pred_y[i], X_test[i])
    print()
```
