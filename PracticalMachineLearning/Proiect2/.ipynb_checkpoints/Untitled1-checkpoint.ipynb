```python
from sklearn.datasets import load_digits
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
%config InlineBackend.figure_format='retina'
```

```python
digits = load_digits()
print(digits.data.shape)

column_names = []
for i in range(160):
    column_names.append(str(i+1))
csv = pd.read_csv("https://raw.githubusercontent.com/bioramble/pca/master/dog.csv", names=column_names[:80])
csv2 = pd.read_csv("https://raw.githubusercontent.com/bioramble/pca/master/cat.csv", names=column_names[80:])
result = pd.concat([csv, csv2], axis=1)

dog = csv
cat = csv2
pet = result

pet = pet.transpose()
# pet
```

```python
pet_np = np.array(pet)
pet_np.shape
```

```python
kmeans = KMeans(n_clusters=2, random_state=0)
clusters = kmeans.fit_predict(pet_np)
kmeans.cluster_centers_.shape
```

```python
fig, ax = plt.subplots(1, 2, figsize=(8, 3))
centers = kmeans.cluster_centers_.reshape(2, 64, 64)
for axi, center in zip(ax.flat, centers):
    axi.set(xticks=[], yticks=[])
    axi.imshow(center.transpose(), interpolation='nearest', cmap=plt.cm.binary)
```

```python
k=1
row=12
col=7

fig=plt.figure(figsize=(20, 20))
for i in args_t:
    img = pet_np[i].reshape(64,64)
    img = np.transpose(img)
    fig.add_subplot(row, col, k)
    k=k+1
    plt.title("Cluster 1 "+str(i))
    plt.imshow(img, cmap="gray")
plt.show()
```

```python
fig=plt.figure(figsize=(20, 20))
columns = 10
rows = 8
for i in range(0, len(args_t)-1):
    img = pet_np[i].reshape(64,64)
    img = np.transpose(img)
    fig.add_subplot(rows, columns, i+1)
    plt.title(clusters[i])
    plt.imshow(img, cmap="gray")
plt.show()
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(digits.target, labels)
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns; sns.set()
mat = confusion_matrix(digits.target, labels)
sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False,
            xticklabels=digits.target_names,
            yticklabels=digits.target_names)
plt.xlabel('true label')
plt.ylabel('predicted label');
```
