```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
%config InlineBackend.figure_format='retina'
```

```python
column_names = []
for i in range(160):
    column_names.append(str(i+1))
dog = pd.read_csv("https://raw.githubusercontent.com/bioramble/pca/master/dog.csv", names=column_names[:80])
cat = pd.read_csv("https://raw.githubusercontent.com/bioramble/pca/master/cat.csv", names=column_names[80:])
pet = pd.concat([dog, cat], axis=1)
```

```python
dog_np = np.array(dog)
dog_t = np.transpose(dog_np)

cat_np = np.array(cat)
cat_t = np.transpose(cat_np)

pet_np = np.array(pet)
pet_t = np.transpose(pet_np)
```

```python
#eliminate common feature from all images
aux = np.transpose(pet_t)
avg_face_vector = aux.mean(axis=1)
print(avg_face_vector.shape)
print(aux.shape)
avg_face_vector = avg_face_vector.reshape(aux.shape[0], 1)
plt.imshow(avg_face_vector.reshape(64,64).transpose(), cmap="gray")
plt.show()
normalized_face_vector = aux - avg_face_vector
normalized_face_vector[normalized_face_vector < 0] = 0
```

```python
# fig=plt.figure(figsize=(20, 20))
# columns = 10
# rows = 16
# for i in range(0, 160):
#     img = normalized_face_vector[i].reshape(64,64)
#     img = np.transpose(img)
#     fig.add_subplot(rows, columns, i+1)
#     plt.imshow(img, cmap="gray")
# plt.show()
```

```python
covariance_matrix = np.cov(np.transpose(normalized_face_vector))
```

```python
eigen_values, eigen_vectors = np.linalg.eig(covariance_matrix)
```

```python
k=900
eigen_vectors = np.sort(eigen_vectors)
k_eigen_vectors = eigen_vectors[0:k, :]
```

```python
eigen_faces = k_eigen_vectors.dot(normalized_face_vector.T)
```

```python
weights = (normalized_face_vector.T).dot(eigen_faces.T)
```

```python
len(normalized_face_vector[0])
```

```python
test_img = pet_t[60].reshape(len(pet_t[0]),1)
test_normalized_face_vector = test_img - avg_face_vector
test_normalized_face_vector[test_normalized_face_vector < 0] = 0
plt.imshow(test_normalized_face_vector.reshape(64,64).transpose(), cmap="gray")
test_weight = (test_normalized_face_vector.T).dot(eigen_faces.T)
index = np.argmin(np.linalg.norm(test_weight - weights, axis=1))
plt.title(index)
plt.show()
```

```python
for i in range(160):
    test_img = pet_t[i].reshape(len(pet_t[i]),1)
    test_normalized_face_vector = test_img - avg_face_vector
    test_normalized_face_vector[test_normalized_face_vector < 0] = 0
    test_weight = (test_normalized_face_vector.T).dot(eigen_faces.T)
    index = np.argmin(np.linalg.norm(test_weight - weights, axis=1))
    if i!=index:
        print("diff")
```
